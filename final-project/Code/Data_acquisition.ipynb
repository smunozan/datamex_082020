```python
# generamos lista de ciudades a las que vamos a scrapear
rand = []
while len(rand)<100:
    r=random.randint(1,1340)
    if r not in rand:
        rand.append(r)
```

```python
#scroll al final de la página
n_ciudades = len(navegador.find_elements_by_tag_name('li'))

while n_ciudades<max(rand):
    navegador.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(4)
    n_ciudades = len(navegador.find_elements_by_tag_name('li'))
```

```python
#variable donde se van a guardar los usuarios
people = []
```

```python
#scraping de todos los viajes de personas

#loop para generar la información de cada ciudad
for i in tqdm(rand):
    try:
        #seleccionamos una ciudad
        cada_ciudad = navegador.find_element_by_xpath(f'/html/body/div[7]/ul/li[{i}]')

        #dar click a la ciudad
        cada_ciudad.click()
        time.sleep(2)

        #tab de people
        people_tab = navegador.find_element_by_xpath('/html/body/div[32]/div[3]/div[2]/div[4]/ul/h3[8]')
        people_tab.click()

        time.sleep(1)
        
        #lista de personas
        people_lst = navegador.find_element_by_class_name('people-here-now')
        people_lst = people_lst.find_elements_by_tag_name('a')
        
        #agregar la información a la lista principal de personas
        for i in people_lst:
            url = i.get_attribute('href')
            if url not in people:
                people.append(url)

        #cerrar ciudad
        close = navegador.find_element_by_xpath('//*[@id="body"]/div[33]/div')
        close.click()

        time.sleep(1)
            
    except:
        pass
```

```python
#para evitar que nos bloquee el api por hacer muchas llamadas, vamos a generar headers diferentes para cada petición
def get_random_ua():
    random_ua = ''
    ua_file = 'agents.txt'
    try:
        with open(ua_file) as f:
            lines = f.readlines()
        if len(lines) > 0:
            prng = np.random.RandomState()
            index = prng.permutation(len(lines) - 1)
            idx = np.asarray(index, dtype=np.integer)[0]
            random_ua = lines[int(idx)].replace("\n","")
    except Exception as ex:
        print('Exception in random_ua')
        print(str(ex))
    finally:
        return random_ua
```

```python
trips_lst =[]
```

```python
%%time

for x in tqdm(people[10:]):
    try:
        url = x
        user_agent = get_random_ua()
        headers = {'user-agent': user_agent}
        response = requests.get(url, headers = headers)

        sopa = bs(response.content, "html.parser")
        viajes = sopa.find_all('tr',{'class':'trip'})
        user = url.split('/@')[1]

        for i in viajes:
            if i.find('td',{'class':'image'}).find('a')!=None:

                duracion = int(i.find('td',{'class':''})['title'].split(' ')[0])
                ciudad = i.find('td',{'class':'name'}).find('h2').text
                pais = i.find('td',{'class':'country'}).text

                data = {'user':user,'City':ciudad,'Country':pais,'Time (days)':duracion}
                trips_lst.append(data)
        time.sleep(0.5)
    except:
        pass
```

```python
trips_df = pd.DataFrame(trips_lst)
```

```python
cities_user = pd.DataFrame(trips_lst).groupby(['user','City','Country']).sum('Time (days)').reset_index()
cities_user.to_csv('../data/user_cities.csv')
```
