```python
#función para generar el match de las ciudades en una sola ejecución
def match_user_input(lst):
    
    #Importar librerias necesarias
    import pandas as pd
    from fuzzywuzzy import process
    
    #llamamos el csv con los datos
    cities_lst = pd.read_csv('../data/cities_for_user_input.csv')
    
    #función que retorna las ciudades foneticamente más parecidas a las que ingresa el usuario
    def find_best_match(misspelled):
        lista = [x.lower() for x in list(cities_lst["City"])]
        closest, ratio = process.extractOne(misspelled, lista)
        indice = lista.index(closest)
        city_country = cities_lst.loc[indice,'City-Country_EN']
        #print(closest,ratio)
        return city_country
    
    #lista con las ciudades finales
    input_user = []
    
    #iteramos la lista del usuario
    for i in lst:
        input_user.append(find_best_match(i))
        
    return input_user
```

```python
#prueba de la función
prueba = ['isl galapago','La paz mx','La paz bolivia','Venecia, it','Lisboa','Cdmx mejico','koh tao', 'madrid españa', 'cuenca']

input_cities = match_user_input(prueba)
input_cities
```
